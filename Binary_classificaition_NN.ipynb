```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("drive/My Drive/tumor_data.csv")
```

```python
data
```

```python
data = data.drop(columns=["id","Unnamed: 32"],axis =1)
```

```python
data['diagnosis'].replace(to_replace = ['B','M'], value = [0,1],inplace = True) 
```

```python
data
```

```python
data = np.array(data)
```

```python
np.random.shuffle(data)
```

```python
data
```

```python
Y = data[:,0]
```

```python
Y
```

```python
X = data[:,1:]
```

```python
X
```

```python
X.shape
```

```python
X_dash = (X-np.mean(X,axis=0))/np.std(X,axis=0)
```

```python
X_dash
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,X_test,Y,Y_test= train_test_split(X_dash,Y,test_size=0.1, random_state=42)
```

```python
print(X.shape)
print(X_test.shape)
print(Y.shape)
print(Y_test.shape)
```

```python
X_train,X_cv,Y_train,Y_cv= train_test_split(X,Y,test_size=0.15, random_state=42)
```

```python
print(X_train.shape)
print(X_cv.shape)
print(Y_train.shape)
print(Y_cv.shape)
```

```python
from keras.layers import InputLayer,Dense, BatchNormalization
from keras.models import Sequential
```

```python
from keras.regularizers import l2
from keras.optimizers import Adam
```

```python
def build_network():

  network = Sequential()
  
  network.add(InputLayer(input_shape=(X_train.shape[1],),))
 
  #kernel_initializer = connection matrix of theta's
  network.add(Dense(units=X_train.shape[1],activation="relu",kernel_initializer="glorot_normal"))
  
  network.add(BatchNormalization())
  
  # network.add(Dense(units=X_train.shape[1],activation="relu",kernel_initializer="glorot_normal",kernel_regularizer=l2()))
  
  # network.add(BatchNormalization())
  
  network.add(Dense(units=1,activation="sigmoid", kernel_initializer="glorot_normal"))

  return network
```

```python
network = build_network()
```

```python
network.compile(optimizer="adam",loss="binary_crossentropy",metrics = ["accuracy"])
```

```python
performance_history = network.fit(x=X_train,y=Y_train,batch_size=5,epochs=50 ,validation_data=(X_cv,Y_cv))
```

```python
network.summary()
```

```python
network.evaluate(x=X_test, y=Y_test)
```
